**Addition Notes on Concept Sets**: Some concept sets already exists such as Demographics and Zip Code Socioeconomic Status. However, I needed to create concept sets for the specific phenotypes that I wanted to pull from the electronic health records. I found that it was best to create each related concept set separately, hence the multiple notebooks. I created the concept sets using the following steps

    Begin as above to extract a Dataset
    Rather than selecting a concept set, click on the plus sign. This will allow you to create a concept set
    Click on the type of phenptype data you need
    Use codes or search terms to get a list of potential data types
    You can click on the heirarchy to click on individual components if you don't want all component
    Click on the green plus sign to add to the concept set
    When you have all the concepts, click on Finish & Review
    Then click Save Concept Set
   
You can edit the concept sets after the fact, as well. 
Currently the concept sets I created are not in the production workspace since I created them in the pre-production workspace. You can refer to the supplemental table of Rosenthal et. al. (https://www.medrxiv.org/content/10.1101/2023.10.23.23296753v2.supplementary-material) for the specific codes used to create the concept sets.

**04 CRC case control for phekb CRC algorithm.ipynb**: Using the pheKB algorithm (https://phekb.org/phenotype/colorectal-cancer-crc) to identify cases and controls. Extensive cleaning of the concept data sets was performed to ensure that the correct codes were included and incorrect codes were excluded. The following sets of cases and controls are created.

    case.set1 := All participants who have 1 or more CRC record, but not inherited. (Called all cases) 
    case.set3 := case.set1, running pheKB algorithm, NOT removing UC/Chron's (Called cleaned cases)
    Screened controls
    Unscreened controls. Note: All controls includes screened and unscreened controls


**05 Plink commands**: Shell and PLINK commands to combine ancestry informative SNPs in the 1KG, HGDP and All of Us dataset. Then QC and filter using genotypes at all datasets. Then output into separate reference and AoU files to be used later. The AoU data is sepearted out into 10 discrete files due to memory issues.

**06 Estimate Admixture using method in PMID: 26482676**: Determine list of related and unrelated subsets to use with calculating admixture estimates and PCs. Calculate admixture using each of the ten All of Us ancestry informative SNP data files, given the unrelated reference data (1KG and HGDP) and the contintental ancestry assignments.

**07 Compute PCs of ancestry**: Compute the PCs of ancestry in the unrelated subset of 1KG and HGDP data. Calculate the percent variance explained by the PCs. Project the AoU data onto these PCs of ancestry (ten separate files of participants).

**08 Demographics**: Demographics - count, age distribution, sex distribution. Further analysis of ancestral distribution of the OTH cluster. 

**09 Adjust PRS by PCs or admixture**: Tables with mean and sd of the PRS before and after adjustments. Random sampling from the unscreened controls (Set 1) using distance from the median of (PC1, PC2, PC3, PC4, PC5). Also getting a purely random sample (Set 2) from the unscreened controls for comparison

**10 Compare PRS OR**: Tables of OR (beta, C.I.) for 1 SD change in PRS with and without adjusting for age and sex. Figures for OR and C.I. comparing quantiles to the middle. Figures of the observed vs expected upper percentiles, with C.I.. Some analysis in response to reviewers regarding the ADM group. 

**11 Compare PRS AUC**: Compare the AUC performance of the adjusted PRSs for CRC cleaned case and screened control data with and without adjusting for age and sex. Creating tables with the estimated AUC and bootstrap 95% C.I.s
